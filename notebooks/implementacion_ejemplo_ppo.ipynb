# Ejemplo de implementación: POO

En este `notebook` se explica la utilización del enfoque de la librería utilizando programación orientada a objetos.
> Nota: este notebook fue ejecutado desde la carpeta `src/ant_colony/`

```python
from utils import *
from aco_tsp_oo import *
import time
```

```python
path = '../../datasets/gr17_d_city_distances.txt'
G = read_data(path)
```

Instanciar clase colonia con:
- hormigas: 2
- nodo inicial: 4
- tasa de evaporación: .5 
- alpha: 3
- beta: 3

```python
# instanciar
colonia_2 = colony(G, 
                 init_node=4, 
                 n_ants=2, 
                 max_iter=10, 
                 rho=.5, 
                 alpha=3,
                 beta=3)
```

Resolver problema **TSP** y mostrar los resultados:

```python
start = time.time()
colonia_2.solve_tsp()
end = time.time()
print(f'- La mejor distancia encontrada por la colonia es: {colonia_2.best_dist}')
print(f'- La mejor ruta encontrada por la colonia es: {colonia_2.best_route}')
print(f"- Tiempo total: {(end-start)} segundos.")
```

Graficar ruta y distancia encontrada por el algoritmo:

```python
colonia_2.plot_route()
```
