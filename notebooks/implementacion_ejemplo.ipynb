## _Traveling Salesman Problem_ usando algoritmo _Ant Colony_


### GR17 Dataset 
Para esta implementación se utilizarán datos de 17 ciudades del conjunto de datos **GR17** disponible [aquí](https://people.sc.fsu.edu/~jburkardt/datasets/tsp/tsp.html).

El objetivo utilizar el algoritmo _ant colony_ de nuestra librería para obtenerla mínima distancia posible en el problema TSP.

```python
import sys
sys.path.append('../src/ant_colony')
from src.ant_colony.aco_tsp import *
data_path2 = '../datasets/gr17.tsp'
```

Cargar grafo con los datos y realizar construcciones intermedias

```python
# lectura de datos en formato tsp
problem = tsplib95.load(data_path2)
# convertir a grafo 
G = problem.get_graph()
# matrix numerica para diccionario de distancias
G_num = nx.to_numpy_matrix(G)
# diccionario de distancias
dic_dists = create_dic_dist_from_graph(G)

# inicializacion de atraccion y feromonas
tau = init_ferom(G)
eta = init_atrac(G,dic_dists)
A = atraccion_nodos(G, tau, eta, alpha=1, beta=5)
```

Visualización del grafo

```python
utils.plot_graph(G, m_plot='graph')
```

Ruta encontrada por una hormiga

```python
route, dist = hormiga_recorre(G,dic_dists, A, tau, 0, x_best=[], y_best= float('inf'))
utils.graph_optim_path(G, route, dist)
```

Colonia de hormigas

```python
# antcolony
ruta, dist = ant_colony(G, dic_dists, init=0, ants=3, max_iter=500, verbose=20)
```
